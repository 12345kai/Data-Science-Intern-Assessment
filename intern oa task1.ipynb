# Determine which properties had negative effective rent growth during lease-up.

```python
rent = pd.concat([rent_1, rent_2])
```

```python
months_Effective_Rent = ['Apr-08.1', 'May-08.1', 'Jun-08.1', 'Jul-08.1', 'Aug-08.1', 'Sep-08.1', 'Oct-08.1', 'Nov-08.1', 'Dec-08.1']
for year in range(9, 21):
    year_str = f"{year:02d}"  
    for month in ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']:
        months_Effective_Rent.append(f"{month}-{year_str}.1")
        
months_Effective_Rent
```

```python
effective_rent = pd.merge(market_lease_up, rent, on = 'ProjID')
effective_rent
```

```python
#Instead of just find the negative effective rent growth and then just stop, I want to find the "most" negative effective
#during the lease-up
negative_growth_props = []
for index, row in effective_rent.iterrows():
    # Find the delivery month (first month with data)
    delivery_month = None
    for month in months_Effective_Rent:
        if month in effective_rent.columns and pd.notna(row[month]) and row[month] > 0:
            delivery_month = month
            break
    
    if delivery_month is None:
        continue
    
    # Get the lease-up time for this property
    lease_up_time = row['Lease_Up_Time']
    
    if pd.isna(lease_up_time):
        continue
    
    # Determine the end month (delivery + lease_up_time)
    delivery_index = months_Effective_Rent.index(delivery_month)
    end_index = delivery_index + int(lease_up_time)
    
    # Skip properties where we don't have complete data
    if end_index >= len(months_Effective_Rent):
        continue
    
    # Get relevant months during lease-up period
    lease_up_months = months_Effective_Rent[delivery_index:end_index+1]
    
    # Get first month's rent
    initial_rent = row[delivery_month]
    if pd.isna(initial_rent) or initial_rent <= 0:
        continue
    
    # Check each subsequent month during lease-up for negative growth compared to first month
    had_negative_growth = False
    lowest_rent = initial_rent
    lowest_rent_month = delivery_month
    
    for month in lease_up_months[1:]:  # Skip the first month
        if month in effective_rent.columns and pd.notna(row[month]) and row[month] > 0:
            current_rent = row[month]
            
            # If current rent is lower than initial rent, we found negative growth
            if current_rent < initial_rent and current_rent < lowest_rent:
                had_negative_growth = True
                lowest_rent = current_rent
                lowest_rent_month = month
    
    if had_negative_growth:
        rent_drop_pct = ((lowest_rent - initial_rent) / initial_rent) * 100
        
        negative_growth_props.append({
            'ProjID': row['ProjID'],
            'Name': row['Name'],
            'Market': row['MarketName'],
            'Initial Rent': initial_rent,
            'Initial Month': delivery_month,
            'Lowest Rent': lowest_rent,
            'Lowest Month': lowest_rent_month,
            'Rent Drop %': rent_drop_pct,
            'Lease-Up Time': lease_up_time,
            'First Status': row['First Status']
        })
```

This approach allows the model to identify specific relationships between property grades (A+, B, C, etc.) and lease-up duration without imposing artificial numerical relationships between grade levels. Similarly, creating dummy variables for geographic identifiers like Zip and City enables the model to recognize location-specific patterns in lease-up performance that might be missed with broader geographic aggregations. For example, the model can identify which specific cities or zip codes have historically faster or slower lease-up periods, capturing local market dynamics, demand variations, and competitive landscapes that significantly influence absorption rates. By implementing these dummy variables, the predictive model gains the ability to recognize that a B-grade property in a particular zip code might lease up faster than an A-grade property in a different location due to supply-demand imbalances specific to those submarkets.

# Feature 2

The Average Rent During Lease-Up feature significantly enhances lease-up time prediction models by capturing the actual pricing strategy implemented during the critical stabilization period. By calculating the average effective rent specifically during each property's unique lease-up phase (from delivery to stabilization), this feature provides direct insight into the relationship between pricing decisions and absorption velocity. Properties with higher average rents during lease-up may face different market challenges than more affordably priced competitors, potentially extending the time needed to reach target occupancy. Unlike static property attributes or overall market metrics, this dynamic feature reflects the property's actual competitive positioning during the exact period being modeled, making it particularly relevant for predicting lease-up duration. Models incorporating this tailored rent metric can better quantify the price-versus-time tradeoff that developers face when setting initial rents, enabling more accurate predictions of how pricing strategy influences the speed at which new properties reach stabilized occupancy.

```python
rent = pd.concat([rent_1, rent_2])

months_Effective_Rent = ['Apr-08.1', 'May-08.1', 'Jun-08.1', 'Jul-08.1', 'Aug-08.1', 'Sep-08.1', 'Oct-08.1', 'Nov-08.1', 'Dec-08.1']
for year in range(9, 21):
    year_str = f"{year:02d}"  
    for month in ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']:
        months_Effective_Rent.append(f"{month}-{year_str}.1")
```

```python
def calculate_average_rent_during_leaseup(rent_df, rent_columns, lease_up_df):
    # Create a copy of the rent dataframe
    result = rent_df.copy()
    result['Average_Rent_During_LeaseUp'] = None
    
    # Loop through each property in the lease-up dataframe
    for index, lease_row in lease_up_df.iterrows():
        proj_id = lease_row['ProjID']
        
        # Skip if lease-up time is missing
        if pd.isna(lease_row['Lease_Up_Time']):
            continue
        
        # Get the corresponding row in the rent dataframe
        rent_rows = rent_df[rent_df['ProjID'] == proj_id]
        if rent_rows.empty:
            continue
        
        # Convert delivery date format (e.g., 'Apr-19') to rent column format (e.g., 'Apr-19.1')
        delivery_date = lease_row['Delivery Date']  # Get the delivery date
        delivery_month = f"{delivery_date}.1"  # Add the .1 suffix for rent columns
        
        # If the converted delivery month is not in our rent columns, skip
        if delivery_month not in rent_columns:
            continue
        
        # Calculate end of lease-up period
        lease_up_time = int(lease_row['Lease_Up_Time'])
        delivery_index = rent_columns.index(delivery_month)
        end_index = min(delivery_index + lease_up_time, len(rent_columns) - 1)
        
        # Get only the months during the lease-up period
        lease_up_months = rent_columns[delivery_index:end_index+1]
        
        # Filter to only include months that exist in the rent dataframe
        valid_months = [m for m in lease_up_months if m in rent_df.columns]
        
        if not valid_months:
            continue
        
        # Calculate average rent during ONLY the lease-up period
        # Only include columns with actual data
        rent_data = rent_rows[valid_months]
        
        # Calculate mean for each row, ignoring NaN values
        avg_rents = rent_data.mean(axis=1, skipna=True)
        
        # Update the result dataframe for each matched row
        for idx in rent_rows.index:
            result.loc[idx, 'Average_Rent_During_LeaseUp'] = avg_rents.loc[idx]
    
    return result

# Apply the function
df_with_avg_rent = calculate_average_rent_during_leaseup(rent, months_Effective_Rent, combine_data)
df_with_avg_rent
```

# Feature 3

The Individual Competitive Position feature engineering approach creates a powerful predictor for lease-up time by quantifying how each property's key attributes compare to market averages. By calculating normalized differences between a property's age (YearBuilt), unit size (AreaPerUnit), and total units (Quantity) against overall portfolio averages, the method creates percentage-based metrics that reveal a property's relative advantages or disadvantages. These metrics are then combined into a weighted Individual_Competitive_Score that prioritizes newer construction (50% weight), larger units (30% weight), and property size (20% weight). This approach significantly enhances lease-up time prediction by capturing how market participants actually evaluate properties - not in isolation, but relative to alternatives. Properties scoring higher on this index (newer, larger units, appropriate scale) would likely experience accelerated lease-up as they offer superior value propositions to prospective tenants compared to their competition, making this feature particularly valuable for anticipating which properties will stabilize more quickly regardless of market location.

```python
def create_individual_competitive_position(df):
    # Calculate overall averages across all properties
    overall_avg_year = df['YearBuilt'].mean()
    overall_avg_area = df['AreaPerUnit'].mean()
    overall_avg_quantity = df['Quantity'].mean()
    
    # Create comparative metrics for each property individually
    result = df.copy()
    result['YearBuilt_vs_Overall'] = result['YearBuilt'] - overall_avg_year
    result['AreaPerUnit_vs_Overall'] = (result['AreaPerUnit'] / overall_avg_area) - 1
    result['Quantity_vs_Overall'] = (result['Quantity'] / overall_avg_quantity) - 1
    
    # Create individual competitive position score
    result['Individual_Competitive_Score'] = (
        result['YearBuilt_vs_Overall'] * 0.5 +
        result['AreaPerUnit_vs_Overall'] * 0.3 +
        result['Quantity_vs_Overall'] * 0.2
    )
    
    return result
```

# Feature 4

The Average Concession During Lease-Up feature enhances lease-up time prediction models by capturing a critical economic factor that directly influences leasing velocity. By calculating the average concession percentage offered specifically during each property's unique lease-up period, this feature provides insight into the pricing strategy and market competitiveness at the exact time when reaching stabilized occupancy is the primary goal. Properties with higher average concessions during lease-up may reflect challenging market conditions, pricing misalignment, or property issues that could extend the time needed to reach target occupancy. This metric serves as a more relevant predictor than overall or current concessions because it isolates the specific pricing adjustments made during the crucial stabilization phase. Models incorporating this feature can better understand the relationship between price incentives and absorption rates, enabling more accurate predictions of how long new properties will take to stabilize based on their concession strategies during the initial lease-up period.

```python
def calculate_average_concession_during_leaseup(concession_df, concession_columns, lease_up_df):
    # Create a copy of the concession dataframe
    result = concession_df.copy()
    result['Average_Concession_During_LeaseUp'] = None
    
    # Loop through each property in the lease-up dataframe
    for index, lease_row in lease_up_df.iterrows():
        proj_id = lease_row['ProjID']
        
        # Skip if lease-up time is missing
        if pd.isna(lease_row['Lease_Up_Time']):
            continue
        
        # Get the corresponding rows in the concession dataframe
        concession_rows = concession_df[concession_df['ProjID'] == proj_id]
        if concession_rows.empty:
            continue
        
        # Convert delivery date format (e.g., 'Apr-19') to concession column format
        delivery_date = lease_row['Delivery Date']
        
        # Determine which column format to use based on your columns
        if f"{delivery_date}.2" in concession_columns:
            delivery_month = f"{delivery_date}.2"  # For months_occupancy format
        elif delivery_date in concession_columns:
            delivery_month = delivery_date  # For months_regular format
        else:
            continue  # Skip if we can't find the delivery month
        
        # Calculate end of lease-up period
        lease_up_time = int(lease_row['Lease_Up_Time'])
        
        if delivery_month in concession_columns:
            delivery_index = concession_columns.index(delivery_month)
            end_index = min(delivery_index + lease_up_time, len(concession_columns) - 1)
            
            # Get only the months during the lease-up period
            lease_up_months = concession_columns[delivery_index:end_index+1]
            
            # Filter to only include months that exist in the concession dataframe
            valid_months = [m for m in lease_up_months if m in concession_df.columns]
            
            if valid_months:
                # Calculate average concession during ONLY the lease-up period
                concession_data = concession_rows[valid_months]
                
                # Calculate mean for each row, ignoring NaN values
                avg_concessions = concession_data.mean(axis=1, skipna=True)
                
                # Update the result dataframe for each matched row
                for idx in concession_rows.index:
                    result.loc[idx, 'Average_Concession_During_LeaseUp'] = avg_concessions.loc[idx]
    
    return result

# Apply the function for months_occupancy format
occupancy_result = calculate_average_concession_during_leaseup(occ_df, months_occupancy, combine_data)
```

# Feature 5

The Effective_Age feature enhances lease-up time prediction by providing a more accurate representation of a property's true condition than simple chronological age. By using the most recent date between original construction (YearBuilt) and renovation (Rehab1), this metric captures whether a property presents as new or recently updated to prospective tenants. This distinction is crucial because renovated older properties often lease up at rates more similar to newer properties than their unrenovated counterparts of the same vintage. The Effective_Age feature efficiently handles missing renovation data by defaulting to the original build year while still capturing the valuable renovation information when available. In predictive models, this single metric provides more explanatory power than separate build and renovation variables, allowing the model to more accurately forecast how a property's perceived age and condition will influence its absorption rate and time to stabilization.

```python
def create_effective_property_age(df):
    # Get current year
    current_year = 2025  # Using the current year
    
    # Create a copy to avoid modifying the original
    result = df.copy()
    
    # Create Effective_Age feature that uses the most recent year between YearBuilt and Rehab1
    result['Effective_Age'] = current_year - result['YearBuilt']
    
    # Where Rehab1 is available, use it instead as it represents a more recent condition
    mask = pd.notna(result['Rehab1'])
    result.loc[mask, 'Effective_Age'] = current_year - result.loc[mask, 'Rehab1']
    
    return result
```

# Embedding Application

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
```

```python
numerical_features = [
    'Lease_Up_Time', 'Latitude', 'Longitude', 'YearBuilt', 
    'Quantity', 'AreaPerUnit', 'Average_Rent_During_LeaseUp',
    'YearBuilt_vs_Overall', 'AreaPerUnit_vs_Overall', 'Quantity_vs_Overall',
    'Individual_Competitive_Score', 'Average_Concession_During_LeaseUp', 'Effective_Age'
]

numerical_features = [col for col in numerical_features if col in combine_data.columns]

grade_features = [col for col in combine_data.columns if col.startswith('Initial_Property_Grade_')]
market_features = [col for col in combine_data.columns if col.startswith('MarketName_')]
state_features = [col for col in combine_data.columns if col.startswith('State_')]
status_features = [col for col in combine_data.columns if col.startswith('First_Status_')]

embedding_features = numerical_features + grade_features + market_features + state_features + status_features
```

```python
X = combine_data[embedding_features].fillna(combine_data[embedding_features].mean())

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA for dimensionality reduction before clustering
pca = PCA(n_components=min(10, len(embedding_features)))
X_pca = pca.fit_transform(X_scaled)
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(X_pca)

# Add cluster labels to the dataframe
combine_data['Cluster'] = clusters
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.manifold import TSNE

kmeans_silhouette = silhouette_score(X_pca, clusters)
print(f"KMeans Silhouette Score (5 clusters): {kmeans_silhouette:.3f}")
```

# Not good, we will try another cluster method. 

```python
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_pca)

# Create a plot
plt.figure(figsize=(12, 8))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=clusters, cmap='viridis', alpha=0.7)
plt.colorbar(scatter, label='Cluster')
plt.title('Property Clusters based on t-SNE Visualization')
plt.xlabel('t-SNE Feature 1')
plt.ylabel('t-SNE Feature 2')
```

```python
import numpy as np
def optimize_dbscan(X, eps_range, min_samples_range):
    best_score = -1
    best_eps = None
    best_min_samples = None
    best_n_clusters = 0
    results = []
    
    for eps in eps_range:
        for min_samples in min_samples_range:
            # Fit DBSCAN
            dbscan = DBSCAN(eps=eps, min_samples=min_samples)
            labels = dbscan.fit_predict(X)
            
            # Count clusters (excluding noise)
            n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
            n_noise = list(labels).count(-1)
            
            # Only calculate silhouette if we have at least 2 clusters and not all points are noise
            if n_clusters >= 2 and n_noise < len(X):
                # Filter out noise points for silhouette calculation
                non_noise_idx = labels != -1
                if sum(non_noise_idx) > 1:
                    try:
                        score = silhouette_score(X[non_noise_idx], labels[non_noise_idx])
                        results.append((eps, min_samples, n_clusters, n_noise, score))
                        
                        if score > best_score:
                            best_score = score
                            best_eps = eps
                            best_min_samples = min_samples
                            best_n_clusters = n_clusters
                    except:
                        pass
    
    return best_eps, best_min_samples, best_n_clusters, best_score, results

# Define parameter ranges to test
eps_range = np.arange(0.5, 3.1, 0.2)
min_samples_range = [3, 5, 7, 10, 15]

# Find optimal parameters
best_eps, best_min_samples, best_n_clusters, best_score, results = optimize_dbscan(X_pca, eps_range, min_samples_range)

print(f"Best parameters: eps={best_eps}, min_samples={best_min_samples}")
print(f"Number of clusters: {best_n_clusters}")
print(f"Best silhouette score: {best_score:.3f}")
```

```python
# DBSCAN with parameters to reduce noise points
dbscan = DBSCAN(eps=0.7, min_samples=3)  # Increased eps, reduced min_samples
dbscan_clusters = dbscan.fit_predict(X_pca)
combine_data['DBSCAN_Cluster'] = dbscan_clusters

# Count number of clusters and noise points
n_dbscan_clusters = len(set(dbscan_clusters)) - (1 if -1 in dbscan_clusters else 0)
n_noise_points = list(dbscan_clusters).count(-1)
print(f"DBSCAN found {n_dbscan_clusters} clusters and {n_noise_points} noise points")

# Calculate silhouette score for non-noise points
if n_dbscan_clusters > 1:
    non_noise_idx = dbscan_clusters != -1
    if sum(non_noise_idx) > 1:
        dbscan_silhouette = silhouette_score(X_pca[non_noise_idx], dbscan_clusters[non_noise_idx])
        print(f"DBSCAN Silhouette Score (excluding noise): {dbscan_silhouette:.3f}")

# Visualize with t-SNE
plt.figure(figsize=(12, 8))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=dbscan_clusters, cmap='viridis', alpha=0.7)
plt.colorbar(scatter, label='Cluster')
plt.title(f'DBSCAN Clustering (eps=1.2, min_samples=2): {n_dbscan_clusters} clusters, {n_noise_points} noise points')
plt.xlabel('t-SNE Feature 1')
plt.ylabel('t-SNE Feature 2')
plt.show()
```

# eps=0.7, min_samples=3 get high silhouette score, but based on the plot the interpretation is poor. So there is a trade-off..
